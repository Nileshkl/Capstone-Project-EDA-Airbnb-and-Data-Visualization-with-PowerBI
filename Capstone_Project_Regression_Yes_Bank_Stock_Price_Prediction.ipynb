#### Chart - 2 Dependent Variable Distribution Analysis

````python
# Chart - 2 visualization code
plt.figure(figsize=(8,5))
sns.distplot(df_data['Close'],color='teal')
plt.title('Distribution of Dependent Variable')
plt.xlabel('Closing Price')
plt.axvline(df_data['Close'].mean(),color='Yellow')
plt.axvline(df_data['Close'].median(),color='Red',linestyle='dashed')
plt.show()
````

- Utilizing a distplot for examining the distribution of the dependent variable.
- The visual representation indicates positive skewness, particularly skewed to the right.
- This skewness suggests inherent volatility in the closing price over time.
- The impact of the 2018 fraud case is evident in this skewness, emphasizing the need for data transformation techniques.
- Appropriate transformations can address the skewness issue and potentially lead to more accurate predictions for the closing price.
- These predictions take into account both historical trends and the impact of significant events like the 2018 fraud incident.

#### Chart - 3: Transformed Dependent Variable Distribution Analysis

````python
# Chart - 3 visualization code
# Applying log transformation
plt.figure(figsize=(8,5))
sns.distplot(np.log10(df_data['Close']),color="teal")
plt.title('Transformed Dependent Variable Distribution Analysis')
plt.xlabel('closing price')


plt.axvline(np.log10(df_data['Close']).mean(),color='yellow')
plt.axvline(np.log10(df_data['Close']).median(),color='red',linestyle='dashed')
plt.show()
````

- The application of a log transformation has significantly altered the characteristics of the dependent variable's distribution.
- Post-transformation, the distribution closely resembles a normal distribution, which is advantageous for modeling.
- The mean and median values converge, indicating symmetry in the data.
- The proximity of these central tendency measures suggests a balanced distribution.
- Achieving normality simplifies the modeling process, aligns with statistical assumptions, and enables the development of more reliable and accurate predictive models.
